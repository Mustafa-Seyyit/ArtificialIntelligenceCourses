# 3 Boyutlu Matris

```python
n, m, c = 3, 3, 3

#3b dizi
image = [[[0]*m for _ in range(n)] for _ in range(c)]

#atama
#image[0][0][0] = 255
#image[1][1][1] = 255
#image[2][2][2] = 255

for i in range(n):
    image[i][i][i] = int(input(f"matris[{i}][{i}][{i}] i√ßin deper giriniz (0 - 255)"))

#gezinme
for derinlik in range(c):
    for satir in range(n):
        for sutun in range(m):
            print(f"{image[satir][sutun][derinlik]:4}", end="")
        print()
    print()    
```
