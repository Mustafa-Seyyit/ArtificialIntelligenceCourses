```python
exp = input("Bir cümle girininiz: ")
harf = input("Aranan harfi giriniz: ")

sayac = 0

for char in exp:
    if char.lower() == harf.lower():
        sayac += 1

print(f"{harf} harfi {sayac} kez ilgili ifadede geçiyor.")
```
