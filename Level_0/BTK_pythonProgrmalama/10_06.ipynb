# Matrislerde Toplama

```python
def print_matris(m):
    print("- "*len(m[0]*3))

    for i in range(len(m)):
        for j in range(len(m[0])):
            print(f"{m[i][j]:4}", end = "")
        print()
    print()
```

```python
satır = int(input("satır sayısını giriniz: "))
sutun = int(input("sutun sayısını giriniz: "))

m1 = [[0]*sutun for i in range(satır)]
m2 = [[0]*sutun for i in range(satır)]
m3 = [[0]*sutun for i in range(satır)]

print_matris(m1)
print_matris(m1)
print_matris(m1)
```

```python
# değer atama / gezinme / yeni atama

for i in range(satır):
    for j in range(sutun):
        m1[i][j] = (i+1) + (j+1)
        m2[i][j] = (i+1) * (j+1)
        m3[i][j]  = m2[i][j] + m1[i][j]

print_matris(m1)
print_matris(m2)
print_matris(m3)
        
```
