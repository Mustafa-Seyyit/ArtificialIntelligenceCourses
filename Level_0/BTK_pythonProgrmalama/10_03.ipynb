```python
import random

satır = int(input("satır sayısını giriniz: "))
sutun = int(input("sutun sayısını giriniz: "))
matris = []

for i in range(satır):
    satır_listesi = []
    for j in range(sutun):
        satır_listesi.append(random.randint(0, 10))
    matris.append(satır_listesi)

print(matris)
```

```python
for i in range(len(matris)):
    for j in range(len(matris[0])):
        print(f"{matris[i][j]:3}", end = " ")
    print()
```

## Transpoz

```python
transpoze = []

for j in range(len(matris[0])):
    yeni_satir = []
    for i in range(len(matris)):
        yeni_satir.append(matris[i][j])
    transpoze.append(yeni_satir)

print(transpoze)
```

```python
for i in range(len(transpoze)):
    for j in range(len(transpoze[0])):
        print(f"{transpoze[i][j]:4d}", end="")
    print()
```
