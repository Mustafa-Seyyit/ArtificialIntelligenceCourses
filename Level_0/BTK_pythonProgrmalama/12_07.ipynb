# Kelime Analizi

```python
#giriş
text = "BTK Akademi ile Programlama tekrar ediyorum."

#ayırma
word_list = text.split()

#kelime sayısı
word_count = len(word_list)

#harf sayısı
letter_count = 0

for char in text:
    if char.isalpha():
        letter_count += 1

#en uzun kelime
longest_word = ""
for word in word_list:
    if len(word) > len(longest_word):
        longest_word = word

print("kelime saysı: ", word_count)
print("harf sayısı: ", letter_count)
print("en uzun kelime: ", longest_word)
print("en uzun kelimenin uzunluğu: ", len(longest_word))
```
