# Girilen Sayinini Asal Olmasının Kontrolü

```python
n = int(input("Bir sayı giriniz: "))

if n < 2:
    print(f"{n} asal değildir.")
elif n == 2:
    print("2 asal sayıdır.")
elif n % 2 == 0:
    print(f"{n} asal sayı değildir, örneğin 2 ve {n//2} sayılarına bölünüyor.")
else:
    asal = True
    bolen = None
    i = 3
    while i * i <= n:          # sadece sqrt(n)'e kadar
        if n % i == 0:
            asal = False
            bolen = i
            break
        i += 2                 # sadece tek sayılar
    if asal:
        print(f"{n} bir asal sayıdır.")
    else:
        print(f"{n} asal sayı değildir, örneğin {bolen} ve {n//bolen} sayılarına bölünüyor.")
```
