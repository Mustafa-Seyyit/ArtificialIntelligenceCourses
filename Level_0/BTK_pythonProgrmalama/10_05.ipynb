# Alt Üçgen Matris

```python
import random

n = 5
matris = []

for i in range(n):
    yeniSatir = []
    for j in range(n):
        yeniSatir.append(random.randint(0,10))
    matris.append(yeniSatir)
print(matris)
```

```python
for i in range(n):
    for j in range(n):
        print(f"{matris[i][j]:4}", end = "")
    print()
```

```python
for i in range(n):
    for j in range(n):
        if j > i:
            matris[i][j] = 0
```
