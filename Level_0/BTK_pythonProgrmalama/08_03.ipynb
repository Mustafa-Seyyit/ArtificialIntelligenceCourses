# sayi tahmini

```python
import random

tutulan = random.randint(1, 100)
denemeSayisi = 0

while True:
    tahmin = int(input("tahmin giriniz: "))
    denemeSayisi +=1

    if tahmin < tutulan:
        print("daha büyük bir tahminde bulununuz.")
    elif tahmin > tutulan:
        print("daha küçük bir tahminde bulununuz.")
    else:
        print(f"tebrikler {denemeSayisi} tahminde {tutulan} sayısını buldunuz.")
        break
```
