```python
def add(*numbers):
    total = 0
    for n in numbers:
        total += n
    return total

print(add(5))
print(add(5, 3))
print(add(5, 3, 9))
print(add(5, 3, 9, 10))
```

```python
def print_details(**info):
    for key, value in info.items():
        print(f"{key} : {value}")

print_details(name = "Mustafa", age = 22, city = "Tokat")
```

```python
def print_details(name = "Misafir", **info):
    print("name: ", name)
    for key, value in info.items():
        print(f" {key} : {value}")

print_details(name = "Mustafa", age = 22, city = "Tokat")
print_details("Seyit", age = 22, city = "Tokat")
print_details(age = 22, city = "Karabük")
```

```python
def divide(a,b):
    try:
        return a/ b
    except ZeroDivisionError:
        return "Sıfıra Bölme Hatası!"
    except:
        return "Hata!" 

print(divide(10 , 2))
print(divide(10 , 0))
```

```python
def check_value(value):
    try:
        number = int(value)
    except ValueError:
        print("lütfen sayısal bir değer giriniz!")
    else:
        print("girilen sayı: ", number)
    finally:
        print("kontrol sağlandı")

check_value(42)
check_value("merhaba")
```

```python
def factorial(n):
    if n < 0:
        raise ValueError("Negatif sayıların faktöriyeli alınamaz!")
    result = 1
    for i in range(1, n+1):
        result *=i
    return result

print(factorial(5))
factorial(-3)
```
