# İstenilen Sayida Fibonacci Dizisi Elemanını Yazdırma

```python
elemanSayisi = int(input("kaç eleman yazdırmak istiyorsunuz: "))

if elemanSayisi <= 0:
    print("pozitif bir sayi girmelisiniz")
else:
    a, b = 0, 1
    for i in range (elemanSayisi):
        print(f"{a}", end= " ")
        a,b = b, a+b
```
