```python
numbers = list(range(1,7))
matris = [[0, 0, 0], [0, 0, 0]]

p = 0

for i in range(2):
    for j in range(3):
        matris[i][j] = numbers[p]
        p +=1

print(matris)
```

```python
for i in range(len(matris)):
    for j in range(len(matris[0])):
        print(f"{matris[i][j]:4d}", end="")
    print()
```
