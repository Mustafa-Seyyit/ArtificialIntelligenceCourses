# faktöriyel hesaplama

```python
hesaplanan = int(input("faktöriyeli hesaplamak istediğiniz sayiyi girini(pozitif tam sayı olmaı)"))
faktoriyel = 1

if hesaplanan < 0:
    print("negatif sayı girmemelisiniz")
else:
    for i in range(1, hesaplanan + 1):
        faktoriyel *= i
    
print(f"{hesaplanan}!: {faktoriyel}")
    
```
