# Kullanıcının Girdiği Pozitif Sayıların Toplamı

```python
toplam = 0
print("pozitif sayılar girin bitirmek için 0 girin")

while True:
    sayi = int(input("sayi: "))
    if sayi == 0:
        break
    elif sayi > 0:
        toplam += sayi
    else:
        print("sayi yok sayıldı")

print(f"girilen sayılardan pozitif olanlarının toplamı: {toplam}")
```
