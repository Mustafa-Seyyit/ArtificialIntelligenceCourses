# 1' den n' e kadar olan sayiların toplamı

```python
while True:
    try:
        n = int(input("n değerini giriniz: "))
        if n < 0:
            print("lütfen geçerli bir sayi giriniz")
        break
    except:
        print("Hata!")

toplam = 0
for i in range(1, n+1):
    toplam += i

print(f"1 den {n}'e kadar olan tam sayıların toplomı: {toplam}")
```
