```python
#giriş
grades_matrix = [[70,80],
                 [90,85,80],
                 [60,75],
                 [10,20,30]]

#çıkış
avarages = []

for student_grades in grades_matrix:
    avarage = sum(student_grades) / len(student_grades)
    avarages.append(avarage)

print("öğrenci not ortalamaları: ", avarages)
```
