```python
#başlangıç sayı listesi
numbers = [2,4,5,6,7,10,0,19]

#asal sayılar
prime_numbers = []

#listede gezinme
for num in numbers:
    if num < 2:
        continue
    
    is_prime = True
    for i in range(2, int(num**0.5) +1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
            prime_numbers.append(num)    

print("Asal sayilar: ", prime_numbers)
```

```python
def is_prime(number):
    if number < 2:
        return False
    for i in range(2, int(number**0.5) +1):
        if number % i == 0:
            return False
    
    return True


#başlangıç sayı listesi
numbers = [2,4,5,6,7,10,0,19]

#asal sayılar
prime_numbers = []

#listede gezinme
for number in numbers:
    if is_prime(number):
        prime_numbers.append(number)

print("Asal sayılar: ", prime_numbers)
```
