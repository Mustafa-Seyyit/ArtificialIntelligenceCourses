```python
toplam = 0

for i in range(1, 101):
    toplam += i
print(f"birden 100 e kadar olan sayıların toplamı: {toplam}")
```

```python
toplam = 0
limit = int(input("limit değerini giriniz: "))

for i in range(1, limit+1):
    toplam += i
print(f"girdiğniz {limit} değerine kadar olan tam sayıların toplamı: {toplam}")
```

```python
toplam = 0
limit = int(input("limit değerini giriniz: "))

for i in range(1, limit+1):
    if i % 2 == 0:
        continue
    toplam += i

print(f"1 den {limit} değerine kadar olan tek sayıların toplamı: {toplam}")
```

```python
toplam = 0
limit = int(input("limit değerini giriniz: "))

for i in range(1, limit +1):
    if i % 2 == 0:
        toplam += i

print(f"1 den {limit} değerine kadar olan çift sayiların toplamı: {toplam}")
```
