# Matrisin Diagonal Elemanları

```python
n = int(input("boyut için bir n değeri giriniz: "))
c = 1
matris = [[0 for i in range(n)] for j in range(n)]
diag = []

for i in range(n):
    for j in range(n):
        matris[i][j] = c
        c += 1

for i in range(n):
    diag.append(matris[i][i])

print("diyagonel elemanlar: ", diag)
print(f"diagonel elemanların toplamı: {sum(diag)}")
```
