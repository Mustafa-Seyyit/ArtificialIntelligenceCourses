# Girilen Sayiların Ortalamasını Dizide Kullanarak Bulma

```python
sayilar = []

for i in range(5):
    s = int(input("lütfen sayi giriniz:"))
    sayilar.append(s)

toplam = 0
for sayi in sayilar:
    toplam += sayi

ortalama = toplam / len(sayilar)

print(f"girilen sayiların ortalaması: {ortalama}")
```
