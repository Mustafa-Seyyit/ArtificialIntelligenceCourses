# Girilen Sayıları Pozitiflik -  Negatiflik Değerlerine Göre Sınıflandırma

```python
#girişler
numbers = input("sayıları aralarında bir boşluk olacak şekilde girinizi (örneğin:10 20 30 ...): ")
#tip dönüşümü
numberList = [int(num) for num in numbers.split()]

positiveNumbers = []
negativeNumbers = []

#sınıflandırma
for num in numberList:
    if num > 0:
        positiveNumbers.append(num)
    elif num < 0:
        negativeNumbers.append(num)
    else:
        print("sıfır herhangi bir listeye eklenmedi")

#sonuçları gönrüntüleme
print(f"pozitif sayılar: {positiveNumbers}")
print(f"negatif sayılar: {negativeNumbers}")
```
