# Frekans Hesaplama

```python
numbers = [1,2,2,3,1,1]

frequency = {}

for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1

for key, value in frequency.items():
    print(f"{key} : {value} kez")
```

```python
for key in frequency.keys():
    print(key)
```
