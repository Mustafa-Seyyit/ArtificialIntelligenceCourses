```python
isimler = ["mustafa", "fatmanur", "seyyit", "dogan"]

for isim in isimler:
    print(isim)
```

```python
for i in range(len(isimler)):
    print(f"{i}. eleman: {isimler[i].title()} ")
```

```python
for i in range(len(isimler)-1, -1, -1):
    print(f"{i}. eleman: {isimler[i]}")
```

```python
i = 0

while i < len(isimler):
    print(f"{i}. eleman: {isimler[i].title()}")
    i +=1
```

```python
i = len(isimler)-1

while i >= 0:
    print(f"{i}. eleman: {isimler[i].title()}")
    i-=1
```

```python
for index, isim in enumerate(isimler):
    print(f"{index}. eleman: {isim}")
```

```python
for index, isim in enumerate(reversed(isimler)):
    print(f"{index}. eleman: {isim.title()}")
```
