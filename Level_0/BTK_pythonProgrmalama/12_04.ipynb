```python
#title : Yapay Zekanın Geleceği
#url : yapay-zekanin-gelecegi

title = input("Tittle: ").lower()
words = title.split()

#key-value
turkish_map={
    "ç": "c",
    "ğ": "g",
    "ı": "i",
    "ü": "u"
}

clean_list = []

for word in words:
    new_word = ""
    for char in word:
        if char in turkish_map:
            new_word += turkish_map[char]
        else:
            new_word += char
    clean_list.append(new_word)

url = "-".join(clean_list)

print(title)
print(url)
```
