```python
numbers = [4,7,1,9,2]
total = 0
for number in numbers:
    total += number

print("toplam: ", total)
```

```python
def sum_list(numbers):
    """
    bir listedeki elemanları toplar.
    geriye toplam sonucu döner
    
    parametre:
    numbers: numaraları içeren listedir.
    """
    total = 0
    for num in numbers:
        total += num
    return total

print("toplam: ", sum_list(numbers))
```

```python
def recursive_sum_list(numbers):
    """
    Bir listedeki elemanları toplar.
    geriye toplamn sonucu döner
    
    parametre:
    numbers: numaraları içeren listedir.
    """

    if len(numbers) == 0:
        return 0
    return numbers[0] + recursive_sum_list(numbers[1:])

print("toplam: ", recursive_sum_list(numbers))
```
