```python
#başlangıç listesi
numbers = [10,15,30,7,41,55,45]

#hem 3 e hem 5 e bölünebilen sayi listesi
divisible_by_3_and_5 = []

#listede döngü
for num in numbers:
    if num %3 == 0 and num %5 == 0:
        divisible_by_3_and_5.append(num)

#sonuçları göster
print("hem 3 e hem de 5 e bölünebilen sayılar: ", divisible_by_3_and_5)
```
