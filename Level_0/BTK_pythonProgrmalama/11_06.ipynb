# Cümlede En Sık Tekrarlanan Karakterler

```python
#kullanıcıdan cümle alma

sentence = input("Bir cümle giriniz: ")
print(sentence)

#boşlukları kaldır
sentence = sentence.replace(" ", "")
print(sentence)

#frekans bilgi

frequency = {}


for c in sentence:
    if c in frequency:
        frequency[c] +=1
    else:
        frequency[c] = 1

#en çok tekrar eden elemanı bul
most_frequency_char = ' '
max_count = 0

#bütün elemanları key - value dolaşma

for key, value in frequency.items():
    if value > max_count:
        max_count = value
        most_frequency_char = key
    
#sonuçları yazdır 
print(frequency)
print("* "*len(sentence)*2)
print(f"en sık tekrar eden karakter {most_frequency_char} ve tekrar değeri {max_count}.")
```
