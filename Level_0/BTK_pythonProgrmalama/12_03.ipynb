```python
pdfs = []
files = ["rapor.pdf", "resim.jpg", "manzara.png", "dokuman.pdf"]

for file in files:
    new_file = file.split(sep='.')
    if new_file[1] == "pdf":
        pdfs.append(file)
print(pdfs)
```

```python
pdfs = []
for file in files:
    if file.endswith("pdf"):
        pdfs.append(file)

print(pdfs)
```
