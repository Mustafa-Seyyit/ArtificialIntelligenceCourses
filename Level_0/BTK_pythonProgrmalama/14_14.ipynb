```python
numbers = [10,25,37,41,53]
target = 371
found = False

for i in range(len(numbers)):
    if numbers[i] == target:
        print(f"{target} bulundu: indeks numarası: {i}")
        found = True
        break
if not found:
    print(target, "listede bulunamadı.")
```

```python
def linear_search(input_list, target):
    """
    doğrusal arama
    
    Args:
        input_list(list): arama yapılacak liste
        target(item): Hedef değer
        
    Returns:
        int: Aranan elemanın liste içerisindeki indir değeri
    """

    for i in range(len(input_list)):
        if input_list[i] == target:
            return i
    return -1

index = linear_search(numbers, 50)

if index != -1:
    print("hedef bulundu! indeks: ", index)
else:
    print("hedef bulunamadı")
```
