```python
import numpy as np
import matplotlib.pyplot as plt
"""
PROBLEM: 5 adet standart normal sayı üret
Karelerini al
Toplamını hesapla
Bu işlemi 10.000 kez tekrar et
Histogram çiz
"""

tekrar_sayisi = 10000
n = 5

sonuclar = []

for _ in range(tekrar_sayisi):
    sayilar = np.random.normal(0, 1, n)
    kareler_toplami = np.sum(sayilar ** 2)
    sonuclar.append(kareler_toplami)

print("İlk 10 sonuç:", sonuclar[:10])
print("Simülasyon ortalaması:", np.mean(sonuclar))
print("Simülasyon varyansı:", np.var(sonuclar))
```

```python
plt.figure()
plt.hist(sonuclar, bins=10)
plt.title("Histogram")
plt.xlabel("Değer")
plt.ylabel("Frekans")
plt.show()
```
